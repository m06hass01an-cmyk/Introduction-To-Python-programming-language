## Python Keywords

- Python keywords are **reserved words** that have a special meaning in the language.  
- They **cannot** be used as variable names, function names, or identifiers.  
- Examples of Python keywords:



## Examples of Python Keywords in Code

```python
# if, else, and elif (conditional keywords)
x = 10
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is 5")
else:
    print("x is less than 5")

# for and break (loop keywords)
for i in range(5):
    if i == 3:
        break
    print(i)

# def and return (function keywords)
def add(a, b):
    return a + b

result = add(2, 3)
print("Sum =", result)

# import (module keyword)
import math
print("Square root of 16 =", math.sqrt(16))

# True, False, None (constants)
flag = True
value = None
print(flag, value)


## Comments in Python

Python has **two types of comments**:

```python
# Single-line comment example
# Single-line comments start with the hash symbol (#)
print("Hello from single-line comment!")

"""
Multi-line comment example
Multi-line comments are enclosed in triple quotes (''' ... ''' or """ ... """)
They can span multiple lines in Python
"""
print("Hello from multi-line comment!")


## Indentation Rules in Python

- Python uses **indentation (spaces or tabs)** to define blocks of code instead of `{}` like in C/C++ or Java.  
- Indentation is **mandatory**; incorrect indentation causes `IndentationError`.  
- Typically, **4 spaces** are used per indentation level (or 1 tab).  

### Examples:

```python
# Correct indentation
if True:
    print("This is correctly indented")
    print("Another statement in the same block")

# Incorrect indentation (will cause an error)
if True:
print("This will cause IndentationError")


## Python `print()` Function

- The `print()` function is used to **display output** on the screen.  
- You can print **text, numbers, variables, or multiple items**.  

### Examples:

```python
# Print a message
print("Hello, Python!")

# Print a number
x = 10
print(x)

# Print multiple items
name = "Ali"
age = 20
print("Name:", name, "Age:", age)


## Python `input()` Function

- The `input()` function is used to **take input from the user**.  
- It **always returns a string**, so you may need to convert it for numbers.  

### Examples:

```python
# Take a name from the user
name = input("Enter your name: ")
print("Hello,", name)

# Take a number from the user
age = int(input("Enter your age: "))
print("You are", age, "years old")
