# Python if, elif, else Statements

Conditional statements are used to **make decisions** in Python based on conditions.

---

## Syntax

```python
if condition:
    # code block if condition is True
elif another_condition:
    # code block if elif condition is True
else:
    # code block if all conditions are False


````python
#Example 1: Simple if–else
age = 18

if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
````

````python
#Example 2: if–elif–else
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
````

````python
#Example 3: Multiple Conditions
num = 0

if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero")
````

# Nested Conditions in Python

Nested conditions mean using an **if–elif–else statement inside another if statement**.  
They are used when a decision depends on **multiple levels of conditions**.

---

## Syntax

```python
if condition1:
    if condition2:
        # code block
    else:
        # code block
else:
    # code block


````python
#Example 1: Basic Nested if
num = 10

if num > 0:
    if num % 2 == 0:
        print("Positive and Even")
    else:
        print("Positive and Odd")
else:
    print("Number is zero or negative")
````

````python
#Example 2: Login System (Real-life Example)
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Incorrect password")
else:
    print("Username not found")
````

````python
#Example 3: Nested Conditions with Marks
marks = 85

if marks >= 50:
    if marks >= 90:
        print("Grade A")
    else:
        print("Grade B")
else:
    print("Fail")
````
